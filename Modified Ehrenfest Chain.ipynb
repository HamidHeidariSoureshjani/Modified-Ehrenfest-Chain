```python
import random
import matplotlib.pyplot as plt
```

```python
def simulate_marbles_process(d, n):
    # Initial State
    x = random.randint(0, d)  # Initial Number of Marbles in Pot 1
    history = [x]

    for _ in range(n):
        if x == 0:
            # We can only add the bead to Pot 1
            x += random.choice([0, 1])
        elif x == d:
            # We can only remove the bead from Pot 1
            x -= random.choice([0, 1])
        else:
            # Random decision to add or remove marble
            change = random.choice([-1, 0, 1])
            x += change

        history.append(x)

    return history

# Configuration
d = 10  # Number of all Marbles
n = 20  # Number of times of the process
```

```python
# Run and print the results
marbles_history = simulate_marbles_process(d, n)
print(marbles_history)
```

```python
# Plot the results
plt.plot(marbles_history)
plt.title('Changes in the number of marbles in the Pot 1')
plt.xlabel('The number of times of the process')
plt.ylabel('The Number of marbles in Pot 1')
plt.show()
```
